```python
import numpy as np
import matplotlib.pyplot as plt

def linear_gradient(width, height, point1, point2):
    img = np.zeros((height, width, 3), dtype=np.uint8)
    x1, y1, start_color = point1 ; x2, y2 , end_color= point2
    for x in range(width):
        for y in range(height):
            t = ((x - x1) * (x2 - x1) + (y - y1) * (y2 - y1)) / float((x2 - x1)**2 + (y2 - y1)**2)
            t = np.clip(t,0,1)
            img[y, x] = [int(start_color[0] * (1 - t) + end_color[0] * t),
                         int(start_color[1] * (1 - t) + end_color[1] * t),
                         int(start_color[2] * (1 - t) + end_color[2] * t)]

    return img
```

```python
width, height = 400, 400
# point = (x , y , color RGB)
point1 = (75,75,[52, 235, 103])
point2 = (400,400,[183, 52, 235])
grad = linear_gradient(width, height, point1, point2)

plt.imshow(grad.astype(np.uint8))
plt.axis('off')
plt.show()
```
